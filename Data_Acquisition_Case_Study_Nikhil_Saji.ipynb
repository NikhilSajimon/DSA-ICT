```python
import pandas as pd
```

```python
population = pd.read_csv('/content/population.csv')
population
```

```python
population.info()
```

```python
gdp = pd.read_excel('/content/gdp.xlsx')
gdp
```

```python
gdp.info()
```

```python
internet_users = pd.read_json('/content/internet_users.json')
internet_users
```

```python
internet_users.info()
```

```python
literacy_rate = pd.read_xml('/content/literacy_rate.xml')
literacy_rate
```

```python
literacy_rate.info()
```

```python
gdp.rename(columns={'Country':'country'}, inplace=True)
literacy_rate.rename(columns={'name':'country'}, inplace=True)
```

```python
merged_df = pd.merge(population, gdp, on='country', how='inner')
merged_df = pd.merge(merged_df, internet_users, on='country', how='inner')
merged_df = pd.merge(merged_df, literacy_rate, on='country', how='inner')
```

```python
merged_df
```

```python
merged_df['internet_penetration_rate'] = merged_df['internet_users']/merged_df['population'] * 100
```

```python
merged_df
```

```python
# Find countries with highest internet penetration rates
```

```python
merged_df.sort_values(by='internet_penetration_rate', ascending=False).head(3)
```

```python
# Calculate average literacy rate across countries
```

```python
merged_df['literacy_rate'].mean()
```

```python
merged_df[['population', 'GDP', 'internet_users', 'literacy_rate']].corr()
```

```python
# Load into SQLite:
```

```python
# # Create an SQLite database
import sqlite3
conn = sqlite3.connect('country_data.db')

# Load the final merged dataset into a database table
merged_df.to_sql('country_data', conn, index=False, if_exists='replace')
```

```python
pd.read_sql_query('select * from country_data', conn)
```

```python
#  Write and execute SQL queries to extract insights
```

```python
# a. Countries with highest internet penetration
pd.read_sql_query('select * from country_data order by internet_penetration_rate desc limit 3', conn)
```

```python
# b. Average literacy rate
pd.read_sql_query('select avg(literacy_rate) from country_data', conn)
```

```python
# c. Correlation between literacy rate and internet penetration
merged_df['literacy_rate'].corr(merged_df['internet_penetration_rate'])
```

```python
# d. GDP per capita analysis
pd.read_sql_query('select country, GDP/population as GDP_per_capita from country_data order by GDP_per_capita desc', conn)
```

```python
# e. Which countries have a GDP per capita above $10,000?
pd.read_sql_query('select country, GDP/population as GDP_per_capita from country_data where GDP_per_capita > 10000', conn)
```

```python
# f. What is the total population covered in the dataset?
pd.read_sql_query('select sum(population) as total_population from country_data', conn)
```

```python
# g. Which countries have the lowest literacy rates, and how does that impact internet access?

pd.read_sql_query('select country, literacy_rate, internet_penetration_rate from country_data order by literacy_rate limit 5', conn)
```

```python
merged_df['literacy_rate'].corr(merged_df['internet_penetration_rate'])
```

```python
#h. What are the top 5 wealthiest countries by total GDP, and how does that compare
# with population size?

pd.read_sql_query('select country, GDP from country_data order by GDP desc limit 5', conn)
```

```python
merged_df['population'].corr(merged_df['GDP'])
```

```python
#i.  Find countries where internet users exceed 70% of the population
pd.read_sql_query('select country from country_data where internet_users > 0.7 * population', conn)
```

```python
# j. What is the average GDP per capita for countries with internet penetration above 50%?

pd.read_sql_query('''select avg(gdp_per_capita)
                     from (select gdp/population as gdp_per_capita
                           from country_data where internet_penetration_rate > 50)''', conn)
```

```python
# k. How many countries have a literacy rate above 90%, and what is their average internet penetration?

pd.read_sql_query('select country from country_data where literacy_rate > 90', conn)
```

```python
pd.read_sql_query('select avg(internet_penetration_rate) from country_data where literacy_rate > 90', conn)
```
