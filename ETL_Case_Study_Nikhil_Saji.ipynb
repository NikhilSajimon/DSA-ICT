```python
import pandas as pd
import numpy as np
```

```python
# 1. Load the dataset into a DataFrame
```

```python
earthquake_data = pd.read_csv('https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv ')
```

```python
# 2.dentify number of records and unique locations
```

```python
print(f'Records: {len(earthquake_data)}')
print(f"Unique locations: {earthquake_data['place'].nunique()}")
```

```python
# 3.Print top 5 rows and column names
```

```python
print('First five rows: ')
earthquake_data.head()
```

```python
print(f"Column names: {earthquake_data.columns.tolist()}")
```

```python
# 4. Convert time to datetime
```

```python
earthquake_data['time'] = pd.to_datetime(earthquake_data['time'])
```

```python
# 5. Drop records with missing latitude, longitude, mag
```

```python
earthquake_data = earthquake_data.dropna(subset = ['latitude', 'longitude', 'mag'])
```

```python
# 6.  Filter only earthquakes with magnitude >= 4.0
```

```python
earthquake_data[earthquake_data['mag'] > 4]
```

```python
# 7. Add a column day_of_week from time
```

```python
earthquake_data['day_of_week'] = earthquake_data['time'].dt.day_name()
```

```python
# 8. Create a column severity_level based on magnitude:
# ○ <4.0: "Low"
# ○ 4.0 - 6.0: "Moderate"
# ○ 6.0+: "High"
```

```python
conditions = [earthquake_data['mag'] < 4,
              earthquake_data['mag'].between(4, 6),
              earthquake_data['mag'] > 6]
choices = ['Low', 'Moderate', 'High']

earthquake_data['severity_level'] = np.select(conditions, choices, default='Unknown')
```

```python
earthquake_data
```

```python
# 9. Count number of earthquakes per place
```

```python
earthquake_data['place'].value_counts()
```

```python
# 10. Compute average magnitude and max depth per day
```

```python
earthquake_summary = earthquake_data.groupby(earthquake_data['time'].dt.date).agg({'depth':'max', 'mag':'mean'}).reset_index()
earthquake_summary.columns = ['date', 'max_depth', 'average_magnitude']
```

```python
earthquake_summary
```

```python
# 11. Save:
# ○ Cleaned dataset to cleaned_earthquakes.csv
# ○ Summary data to earthquake_summary.csv
```

```python
earthquake_data.to_csv('cleaned_earthquakes.csv')
earthquake_summary.to_csv('earthquake_summary.csv')
```
