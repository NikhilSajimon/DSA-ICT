```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
superstore = pd.read_excel('/content/Sample - Superstore.xls')
```

```python
superstore.head()
```

1. Univariate Analysis

```python
# pie chart of sales by category
```

```python
plt.figure(figsize=(8,5))
category_sales = superstore.groupby('Category')['Sales'].sum()
colors = sns.color_palette('viridis')
plt.pie(category_sales, labels=category_sales.index, autopct='%1.1f%%', colors=colors, shadow=True)
plt.title('Sales Distribution by Category')
plt.axis('equal')
plt.show()
```

```python
# pie chart of profit by region
```

```python
profit_data = superstore.groupby('Region')['Profit'].sum()
colors= sns.color_palette('cubehelix')
plt.pie(profit_data, labels=profit_data.index, autopct='%1.1f%%', colors=colors)
plt.title('Profit by Region')
plt.show()
```

```python
# Profit Distribution
```

```python
plt.hist(superstore['Profit'], bins=[0,50,100,250,500,1000,2000,3000,5000], color='skyblue', edgecolor='black', alpha=0.9)
plt.title('Profit Distribution', fontweight = 'bold')
plt.xlabel('Profit')
plt.ylabel('Frequency')
plt.show()
```

```python
print('''Most of the orders in your data make very little profit.
The chart shows that small profits happen very often, while big profits are rare.
As profit increases, the number of orders drops quickly.''')
```

```python
# Frequency distribution of quantity per order
```

```python
plt.hist(x='Quantity', data=superstore, color='teal', bins=range(1, superstore['Quantity'].max()), edgecolor='k')
plt.title('Frequency distribution of quantity per order')
plt.xlabel('Quantity per order')
plt.ylabel('Frequency')
plt.show()
```

```python
print('''Most customers order small quantities, usually between 2 and 4 items.
The most common order size is 3, and very few orders go above 8 items.''')
```

```python
# 2. Category wise sales
```

```python
plt.bar(superstore['Category'], superstore['Sales'], color='lightcoral', edgecolor='k')
plt.title('Category wise sales', fontweight='bold')
plt.xlabel('Category', fontweight='bold')
plt.ylabel('Sales')
plt.show()
```

```python
print('''Among the three product categories, Technology has the highest sales, followed by Office Supplies, and then Furniture.
''')
```

```python
# Segment wise customers
```

```python
sns.countplot(x = 'Segment', hue = 'Segment', data = superstore, palette='Set1')
plt.title('Segment wise customers', fontweight='bold')
plt.xlabel('Segment', fontweight='bold')
plt.ylabel('Customer count', fontweight='bold')
```

```python
print('''Most customers belong to the Consumer segment, followed by Corporate, and then Home Office.
This means the business mainly serves individual buyers, with fewer customers from corporate or home office setups.''')
```

2. Bivariate Analysis


```python
# Sales-Profit analysis
```

```python
plt.scatter(superstore['Sales'], superstore['Profit'], color='orchid')
plt.title('Sales vs. Profit', fontweight = 'bold')
plt.xlabel('Sales', fontweight= 'bold')
plt.ylabel('Profit', fontweight='bold')
```

```python
print('''Most orders have low sales and low profit, which means small transactions are common.
There are a few outliers with very high sales, but they don’t always bring high profit.
''')
```

```python
# Analysis of sales by region
```

```python
plt.bar(superstore['Region'], superstore['Sales'], color='deepskyblue')
plt.title('Sales by Region', size=12, fontweight='bold')
plt.xlabel('Region', fontweight='bold')
plt.ylabel('Sales', fontweight='bold')
plt.show()
```

```python
print('''The South region has the highest sales, followed by West, then Central, and finally East.
This means most of the business revenue comes from the southern part of the country, while the eastern region contributes the least.
''')
```

```python
# Analysis of monthly sales and profits by months
```

```python
superstore['Months'] = pd.to_datetime(superstore['Order Date']).dt.to_period('M')

monthly_sales = superstore.groupby('Months')[['Sales']].sum().reset_index()
monthly_profit = superstore.groupby('Months')[['Profit']].sum().reset_index()
monthly_sales['Months'] = monthly_sales['Months'].dt.to_timestamp()
monthly_profit['Months'] = monthly_profit['Months'].dt.to_timestamp()

fig, ax = plt.subplots(1, 2, figsize=(16, 8), sharex=True)

fig.suptitle('Monthly sales and profits', size=20, fontweight='bold')

ax[0].plot(monthly_sales['Months'], monthly_sales['Sales'],'o-', c='darkorchid',)
ax[0].set_xlabel('Months', size=12, fontweight='bold')
ax[0].set_ylabel('Sales', size=12, fontweight='bold')
ax[0].grid(True)

ax[1].plot(monthly_profit['Months'], monthly_profit['Profit'], 'o-', c='maroon')
ax[1].set_xlabel('Months', size=12, fontweight='bold')
ax[1].set_ylabel('Profit', size=12, fontweight='bold')
ax[1].grid(True)

plt.tight_layout()
plt.show()
```

```python
print('''Over the years, both sales and profits show an overall upward trend, though profits are more variable with noticeable ups and downs.
Sales grow steadily, while profit spikes and dips suggest that not all high sales months are equally profitable.
This could be due to discounts, product mix, or operational costs.
''')
```

3. Multivariate analysis

```python
# Sales vs profit by category
```

```python
sns.scatterplot(data=superstore, x='Sales', y='Profit', hue='Category', palette='Set1')
plt.title('Sales vs Profit by Category')
plt.show()
```

```python
print('''Different product categories show different patterns.
Technology tends to have higher sales and profit, while Furniture shows more losses and scattered results.
Office Supplies sits somewhere in between.
This means not all categories perform equally—some bring in more profit even with similar sales.''')

```

```python
# Heatmap of profit by region and category
```

```python
pivot = pd.pivot_table(data=superstore, index='Region', columns='Category', values='Profit', aggfunc='sum')
sns.heatmap(pivot, fmt='.1f', annot=True)
plt.title('Profit by Region and Category', fontweight='bold')
plt.show()
```

```python
print('''- Technology is the most profitable category in every region, especially in the West and East.
Office Supplies also perform well, with strong profits in the East and Central.
Furniture shows mixed results—losses in Central, but decent profits elsewhere.
This means the business earns the most from Technology, while Furniture might need attention in certain regions.''')
```
