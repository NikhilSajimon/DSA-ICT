```python
# 1) Import pandas, numpy
```

```python
import pandas as pd
import numpy as np
```

```python
# 2) Load dataset using pd.read_csv()
```

```python
Movies = pd.read_csv('https://raw.githubusercontent.com/laxmimerit/All-CSV-ML-Data-Files-Download/master/IMDB-Movie-Data.csv')
Movies.head()
```

```python
# 3) Display: shape, head, column names, datatypes, missing value
```

```python
print(f'Shape:\n {Movies.shape}\n')
print(f'Head: \n {Movies.head()}\n')
print(f'Columns are: \n {Movies.columns.tolist()}\n')
print(f'Datatypes: \n{Movies.dtypes}\n')
print(f'Missing values per column: \n{Movies.isna().sum()}')
```

```python
# 4) Check and remove duplicates
```

```python
duplicates = Movies[Movies.duplicated()]
print(duplicates)
Movies.drop_duplicates(inplace=True)
duplicates_2 = Movies[Movies.duplicated()]
print(duplicates_2)
```

```python
# 5) Rename columns ("Title" > "Movie_Title")
```

```python
Movies.rename(columns={'Title':'Movie_Title'}, inplace=True)
Movies.head()
```

```python
# 6) Get unique genres
```

```python
uniques = set()
Movies['Genre'].apply(lambda x: uniques.update(x.split(',')))
uniques
```

```python
# 7) Extract number of genres per movie
```

```python
Movies['Genre_count'] = Movies['Genre'].apply(lambda x: len(x.split(',')))
Movies
```

```python
# 8) Handle missing values in 'Revenue (Millions)' and 'Metascore
```

```python
Movies['Revenue (Millions)'].fillna(Movies['Revenue (Millions)'].median())
Movies['Metascore'].fillna(Movies['Metascore'].mean())
```

```python
# 9) Top 5 movies by Rating
```

```python
Movies.sort_values('Rating', ascending=False).head()
```

```python
# 10) Movies with highest/lowest revenue
```

```python
Movies.loc[Movies['Revenue (Millions)'].idxmax()]
```

```python
Movies.loc[Movies['Revenue (Millions)'].idxmin()]
```

```python
# 11) Avg. rating by Director
```

```python
Movies.groupby('Director')['Rating'].mean()
```

```python
# 12) Movies per year
```

```python
Movies.groupby('Year')['Rank'].count()
```

```python
# 13) Movies with rating > 8 & revenue > 100M
```

```python
Movies[(Movies['Rating'] > 8) & (Movies['Revenue (Millions)'] > 100)]
```

```python
# 14) Group by Director: avg Rating, Revenue
```

```python
Movies.groupby('Director').agg({'Rating': 'mean', 'Revenue (Millions)': 'sum'})
```

```python
# 15) Group by Year: count, avg Rating
```

```python
Movies.groupby('Year').agg({'Rank': 'count', 'Rating': 'mean'})
```

```python
# 16) Create a new column Profitability:
# a) Revenue divided by Duration
# b) Find top 10 most profitable movies
```

```python
Movies['Profitability'] = Movies['Revenue (Millions)'] / Movies['Runtime (Minutes)']
Movies.sort_values('Profitability', ascending=False)[['Movie_Title', 'Revenue (Millions)']].head(10)
```

```python
# 17) Create a column that categorizes movies by rating:
# a) Rating < 6: Poor
# b) 6 ≤ Rating < 7.5: Average
# c) Rating ≥ 7.5: Excellent
```

```python
Movies['Rating'] = pd.to_numeric(Movies['Rating'], errors='coerce')
conditions = [Movies['Rating'] < 6, (Movies['Rating'] >= 6) & (Movies['Rating'] < 7.5), Movies['Rating'] >= 7.5]
values = ['Poor', 'Average', 'Excellent']
Movies['Category'] = np.select(conditions, values, default= 'Unclassified')
Movies.head(10)
```

```python
# 18) Create a custom function to score movies based on:
# i) Apply row['Rating'] * 10 + row['Revenue (Millions)'] as High score
# Apply a function to dataset to assign the score
```

```python
def high_score(x):
  return x['Rating'] * 10 + x['Revenue (Millions)']

Movies['High Score'] = Movies.apply(high_score, axis=1)
Movies.head(10)
```
